1.Load The Data

```python
import pandas as pd
df = pd.read_csv("C:/Users/MCST/OneDrive/Desktop/Railwayy.csv") 
df
```

2.Basic Dataset Overview

```python
print(df.info())

print(df.describe())
```

3.Key Statistics to Calculate (KPIs )

```python
# Convert numeric columns
numeric_cols = ["Final_Price","Price","Discount_Amount",
                "Duration_Of_Trip","Duration_Of_Delay",
                "Total_Seats","Booked_Seats"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
df["Occupancy"] = df["Booked_Seats"] / df["Total_Seats"]
df["Price_Diff"] = df["Price"] - df["Final_Price"]
df["Is_Delayed"] = (df["Duration_Of_Delay"] > 0).astype(int)    
```

```python
print("Total Transactions:", len(df))
print("Total Revenue:", df["Final_Price"].sum())
print("Average Final Price:", df["Final_Price"].mean())
print("Average Occupancy Rate:", df["Occupancy"].mean())
print("Delay Rate (%):", df["Is_Delayed"].mean()*100)
print("Average Ticket Price: $", df['Final_Price'].mean())
print("Max Ticket Price: $", df['Final_Price'].max())
print("Min Ticket Price: $", df['Final_Price'].min())
```

 4.Delay buckets

```python
print("\n--- Delay Buckets ---")
delay_bins = pd.cut(df["Duration_Of_Delay"], bins=[-1,0,5,15,30,60,9999],
                    labels=["On Time","0-5m","6-15m","16-30m","31-60m","60m+"])
df["Delay_Bucket"] = delay_bins
print(delay_bins.value_counts())
```

5.Grouped Analysis

```python
df['Duration_Of_Delay'] = pd.to_numeric(df['Duration_Of_Delay'], errors='coerce')
print("Average Delay (minutes):", df['Duration_Of_Delay'].mean())
print("Max Delay (minutes):", df['Duration_Of_Delay'].max())

print("\nAverage Price by Ticket Class:")
print(df.groupby("Ticket Class")["Final_Price"].mean())

print("\nRefund Rate by Delay Bucket:")
print( df.groupby("Delay_Bucket", observed=True)["Refund Request"]
      .value_counts(normalize=True))
print("\nOccupancy by Booking Type:")
print(df.groupby("Booking_Date_Type")["Occupancy"].mean())
```

6.Correlation Checks

```python
print("\nCorrelation with Final Price:")
print(df[["Final_Price","Discount_Amount","Duration_Of_Trip","Occupancy"]].corr())

print("\nCorrelation with Delay Duration:")
print(df[["Duration_Of_Delay","Final_Price","Occupancy","Price_Diff"]].corr())
```

7.Category Analysis

```python
print("\nTop Payment Methods:")
print(df["Payment Method"].value_counts().head())

print("\nTop Ticket Types:")
print(df["Ticket Type"].value_counts().head())

print("\nTicket Class Distribution:")
print(df["Ticket Class"].value_counts())

print("\nTop Railcards Used:")
print(df["Railcard_ Used"].value_counts())
```

8. Delay & Refunds

```python
print("\nDelayed Journeys (%):", (df["Duration_Of_Delay"] > 0).mean()*100)
print("Refund Requests:")
print(df["Refund Request"].value_counts())

print("\nAverage Delay by Reason:")
print(df.groupby("Reason for Delay")["Duration_Of_Delay"].mean())
```

Route Popularity

```python
print("Top 5 Departure Stations:")
print(df['Departure Station'].value_counts().head())

print("Top 5 Arrival Destinations:")
print(df['Arrival Destination'].value_counts().head())
```

 9.Discounts & Pricing

```python
df["Price_Diff"] = df["Price"] - df["Final_Price"]
print("\nAverage Discount:", df["Discount_Amount"].mean())
print("Average Price Difference:", df["Price_Diff"].mean())

print("\nAverage Final Price by Booking Type:")
print(df.groupby("Booking_Date_Type")["Final_Price"].mean())
```
